```python
## map / filter
## iterable (adj)
scores = [82, 78, 81, 66, 50]
```

```python
for score in scores:
  if score >= 80:
    print("pass")
  else:
    print("failed")
```

```python
def grading(score):
  if score >= 80:
    return "passed"
  else:
    return "failed"
```

```python
## map
list(map(grading, scores))
```

```python
## add five to scores
new_scores = []
for score in scores:
   new_scores.append(score+5)
print(new_scores)
```

```python
## add five to scores
list(map(lambda x: x+5, scores))
```

```python
students_score_over_80 = []
for score in scores:
if score >= 80:
  students_score_over_80.append(score)
```

```python
list(filter(lambda x: x>= 80, scores))
```

```python
scores = [ ("fluke", 80), ("b", 78), ("pass", 95) ]

## filter score >= 80
list(filter(lambda x: x[1]>= 80, scores))
```

```python
## module/ library
## homework: class ATM
import random

class ATM:
  def __init__ (self, name, balance):
    self.name = name
    self.balance = balance

  def __str__(self):
    return f"Account name: {self.name}"

  def get_otp(self):
    otp = random.randint(10000, 99999)
    return f"Your Requested OTP: {otp}"
```

```python
my_account = ATM("fluke", 500)
```

```python
print(my_account)
```

```python
## ATM method
my_account.get_otp()
```

```python
my_account.balance
```
