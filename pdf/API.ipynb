```python
import requests
import pandas as pd

sw = []
# loop api 10 species starwar
for i in range(10):
    url = f"https://swapi.info/api/species/{i+1}"
    response = requests.get(url)
    response_json = response.json()
    data = [response_json["name"],
            response_json["designation"],
            response_json["classification"],
            response_json["average_height"],
            response_json["average_lifespan"]]
    sw.append(data)

# create dataframe for response_json
starwars = pd.DataFrame(sw, columns = ["name", "designation", "classification", "average_height", "averrage_lifespan"])
```
