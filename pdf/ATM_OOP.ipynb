```python
import random

class ATM:
    def __init__(self, username, pin_id, bank, amount):
        self.username = username
        self.pin_id = pin_id
        self.bank = bank
        self.amount = amount
        self.is_authenticated = False

    def __str__(self):
        return f"User: {self.username} | Bank: {self.bank} | Balance: {self.amount} THB"

    def login(self, entered_pin):
        if entered_pin == self.pin_id:
            otp = random.randint(1000, 9999)
            print(f"OTP sent to your phone: {otp}")
            verify = int(input("Enter OTP to verify: "))

            if verify == otp:
                self.is_authenticated = True
                print("Login Successful!")
            else:
                print("Invalid OTP. Access Denied.")
        else:
            print("Wrong PIN.")

    def check_balance(self):
        if self.is_authenticated:
            print(f"Your current balance is: {self.amount} THB")
        else:
            print("Please login first.")

    def withdraw(self, money_out):
        if self.is_authenticated:
            if money_out <= self.amount:
                self.amount -= money_out
                print(f"Withdraw: {money_out} THB | Remaining: {self.amount} THB")
            else:
                print("Insufficient balance!")

    def deposit(self, money_in):
        if self.is_authenticated:
            if money_in > 0:
              self.amount += money_in
              print(f"--- Deposit Sucess ---")
              print(f"Added: + {money_in} THB")
              print(f"Current Balance: {self.amount} THB")
            else:
              print("Error: Deposit amount must be positive!")

    def pay_bill(self, bill_name, bill_amount):
        if self.is_authenticated and self.amount >= bill_amount:
            self.amount -= bill_amount
            print(f"Paid {bill_name} bill for {bill_amount} THB. Success!")
        else:
            print("Payment failed: Insufficient funds or not logged in.")

    def order_coffee(self):
        menu = {"Latte": 60, "Espresso": 55, "Cappuccino": 65}
        print("Menu: 1. Latte(60) 2. Espresso(55) 3. Cappuccino(65)")
        choice = input("Select coffee (1-3): ")

        # แปลงตัวเลือกเป็นชื่อกาแฟ
        items = list(menu.keys())
        selected_item = items[int(choice)-1]
        price = menu[selected_item]

        if self.amount >= price:
            self.amount -= price
            print(f"Ordering {selected_item}... Done! Enjoy your coffee.")
        else:
            print("Not enough money for coffee.")
```

```python
my_wallet = ATM("Fluke", "0612", "SCB", 10000)
```

```python
my_wallet.login("0612")
```

```python
print(my_wallet)
```

```python
my_wallet.check_balance()
```

```python
my_wallet.withdraw(2500)
```

```python
my_wallet.pay_bill("Internet dtac", 598)
```
