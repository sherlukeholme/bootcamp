```python
import random

class PizzaGameStore:
    def __init__(self):
        self.menu = {"Hawaiian": 299, "Seafood": 359, "Cheese": 259}
        self.cart = []
        self.total_price = 0
        self.discount = 0

    def order_pizza(self):

      print("\n--- Welcome to fla_luke Pizza Companoon ---")
      print("Menu: Hawaiin (299), Seafood (359), Cheese(259)")

      while True:
        choice = input("what would you like to order ? (Or type 'done' to finish): ").capitalize()
        if choice == 'Done' :
          break

        if choice in self.menu:
          self.cart.append(choice)
          self.total_price += self.menu[choice]
          print(f">> Added {choice}. Total now: {self.total_price} THB")
        else:
          print("!! Sorry, we don't have that menu. Please try again.")

    def play_for_discount(self):
        if not self.cart:
            return

        print("\n--- Play Rock-Paper-Scissors for a 50 THB Discount! ---")
        print("Rule: First to reach 3 wins takes it all!")

        user_win = 0
        com_win = 0
        options = ["rock", "paper", "scissors"]

        while user_win < 3 and com_win < 3:
            user_choice = input(f"\n[You: {user_win} | COM: {com_win}] Choose (rock, paper, scissors): ").lower()

            if user_choice not in options:
                print("Invalid choice! Please try again.")
                continue

            com_choice = random.choice(options)
            print(f"Computer chose: {com_choice}")

            if user_choice == com_choice:
                print("It's a tie!")
            elif (user_choice == "rock" and com_choice == "scissors") or \
                 (user_choice == "paper" and com_choice == "rock") or \
                 (user_choice == "scissors" and com_choice == "paper"):
                user_win += 1
                print("You win this round!")
            else:
                com_win += 1
                print("Computer wins this round!")

        if user_win == 3:
            print("\nCongratulations! You won 3 times. You got a 50 THB discount!")
            self.discount = 50
        else:
            print("\nToo bad! Computer won 3 times. No discount for you.")

    def check_bill(self):
        final_price = self.total_price - self.discount
        print("\n============================")
        print("       RECEIPT SUMMARY       ")
        print("============================")
        print(f"Items: {', '.join(self.cart)}")
        print(f"Subtotal: {self.total_price} THB")
        print(f"Discount: -{self.discount} THB")
        print(f"Total:    {final_price} THB")
        print("============================")
        print("   Thank you for ordering!   ")
```

```python
store = PizzaGameStore()
```

```python
store.order_pizza()
```

```python
store.play_for_discount()
```

```python
store.check_bill()
```
